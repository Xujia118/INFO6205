```python
############################################################
# Recursion.py 
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2021
###########################################################

############################################################
#  All imports here
###########################################################

############################################################
#  class  test factorial
###########################################################    
class Recursion():
    def __init__(self):
        #Nothing can be added here
        pass

    ############################################################
    #          WRITE ALL YOUR PUBLIC FUNCTION BELOW
    ########################################################### 
    def factI(self, n):
        result = 1
        
        while n > 1:
            result *= n
            n -= 1
        
        return result

    def factR(self, n):
        return self._factR(n)

    def printI(self, n, reverse:'bool'):
        if n == 0:
            print(0)

        while n > 0:
            module = n % 10
            print(module)
            n //= 10

        return 
        
    def printR(self, n, reverse:"bool"):
        if n == 0:
            return

        module = n % 10
        print(module)
        self.printR(n // 10, True)
        
        return 

    def reverseI(self, n):
        s = 0

        while n > 0:
            module = n % 10
            s = s * 10 + module
            n //= 10

        return s

    def reverseR(self, n):
        def helper(n, s):
            if n == 0:
                return s
            
            last_digit = n % 10
            s = s * 10 + last_digit
            return helper(n // 10, s)
        
        s = 0
        return helper(n, s)

    def FibI(self, n):
        a, b = 1, 1

        for _ in range(n - 2):
            temp = b
            b = a + b
            a = temp
    
        return b

    def Fib_view(self, n):
        fib = [1, 1]

        for _ in range(n - 2):
            new_item = fib[-1] + fib[-2]
            fib.append(new_item)
        
        return fib
        
    def FibR(self, n):
        if n <= 2:
            return 1
        
        return self.FibR(n - 2) + self.FibR(n - 1)


    ############################################################
    #          WRITE ALL YOUR PRIVATE FUNCTION BELOW
    ########################################################### 

    def _factR(self, n, total=1):
        if n <= 1:
            return total
        
        total *= n
        return self._factR(n - 1, total)
    
    


check = Recursion()
# print(check.reverseI(1982))
# print(check.reverseR(1982))

# check.printI(1986, True)
# check.printR(1986, False)

# print(check.FibI(10))
# print(check.FibR(10))
# print(check.Fib_view(10))

print(check.factR(0))
    
```
